```python
from kaggle_secrets import UserSecretsClient
from openai import OpenAI

API_KEY = UserSecretsClient().get_secret("blablador")

client = OpenAI(
    api_key=API_KEY,
    base_url="https://api.helmholtz-blablador.fz-juelich.de/v1/"
)

models = client.models.list()
print([m.id for m in models.data])
```

```python
!python blablador_translate.py \
  --input dev10.en \
  --output dev10.blablador.nl \
  --source-lang English \
  --target-lang Dutch \
  --model alias-fast \
  --batch-size 1 \
  --progress \
  --print-every 1 \
  --timeout 20 \
  --max-retries 8 \
  --retry-wait 2 \
  --kaggle-secret blablador
```
