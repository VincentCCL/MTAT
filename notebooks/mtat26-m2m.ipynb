# 5.2.4.7 Hands-on M2M-100 Usage

The M2M-100 models are too large to be trained / finetuned in Kaggle, so we skip this step. Conceptually it is similar to finetuning of the other models in the previous sections.

But we can translate with the pretrained model.

## Installing

```python
!wget https://raw.githubusercontent.com/VincentCCL/MTAT/refs/heads/main/code/m2m_translate.py
!pip -q install evaluate sacrebleu
```

## Usage

```python
%run m2m_translate.py \
  --model-dir facebook/m2m100_418M \
  --src-lang en \
  --tgt-lang nl \
  --src-file /kaggle/input/datasets/vincentvandeghinste/tatoeba-en-nl/dev.en \
  --out-file dev.m2m.nl \
  --ref-file /kaggle/input/datasets/vincentvandeghinste/tatoeba-en-nl/dev.nl \
  --metrics bleu,chrf \
  --batch-size 4 \
  --max-src-len 128 \
  --max-gen-len 128 \
  --num-beams 1
```
